## Basic usage

Creation of the front instance, selection of the database and visualisation

```python
import pyvoa.front as pf           # the front end !
pf.setvis('bokeh')                 # using bokeh visualisation
pf.setwhom('owid',reload=False)    # using ourworldindata.org DataBase
```

Simple plot, histogram and map

```python
pf.plot(where='European Union')
pf.hist(where='Asia')
pf.map()
```

## Going more deeply in pyvoa
### Changing the visualisation

```python
pf.listvis()
```

```python
pf.setvis('matplotlib')
pf.plot(where='European union')
```

```python
pf.setvis('bokeh')
pf.plot(where='European union')
```

### Choose the database / the variable to plot

#### databases

```python
pf.listwhom() # list of supported databases
```

```python
pf.setwhom('owid')
```

```python
pf.setwhom('owid',reload=False) # not to retreive data from the web → faster
```

#### variables

```python
pf.listwhich()
```

```python
pf.plot(which='cur_icu_patients')
```

#### pre-treatment of data

```python
pf.listwhat()   # first one is default
```

```python
pf.plot()
```

```python
pf.plot(what='daily')
```

```python
pf.plot(what='weekly')
```

#### Options

```python
pf.listoption()   # default = no option
```

```python
pf.plot()
```

```python
pf.plot(option='sumall')
```

```python
pf.plot(option=['sumall','normalize:pop100']) # % normalisation 
```

#### Location

```python
pf.plot(where=['France','Italy','Esp']) # specifying countries
```

```python
pf.plot(where='European Union') # specifying region
```

### Changing type of graphical output

#### Plots

```python
pf.listplot() # first one is default
```

```python
pf.plot(where='European union',typeofplot='date') # default
```

```python
pf.plot(where='France',what='weekly',which='total_deaths',typeofplot='spiral')   # be carefull, only 1 date can be plotted in spiral mode
```

```python
pf.plot(where='France',what='weekly',which='total_deaths',typeofplot='yearly')    # be carefull, only 1 date can be plotted in yearly mode
```

```python
pf.plot(where=['Italy','France','Germany'],typeofplot='compare')   # needs at least 2 locations
```

```python
pf.plot(which=['total_deaths','total_cases'],typeofplot='versus')   # needs to variables
```

#### Histograms

```python
pf.listhist()   # first is default
```

```python
pf.hist(where='South America')
```

```python
pf.hist(which='total_people_vaccinated',where='Europe',typeofhist='value')
```

```python
pf.hist(which='total_people_vaccinated',where='Europe',typeofhist='pie')
```

#### Maps

```python
pf.listmap()    # first one is default
```

```python
pf.setwhom('spf',reload=False)   # for the demo…
```

```python
pf.map(typeofmap='not dense')
```

```python
pf.map(typeofmap='dense')
```
