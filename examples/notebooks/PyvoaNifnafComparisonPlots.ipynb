## Evolution du nombre de patients en réanimation en Ile de France depuis le 22/03/2020 (première vague) [(padlet page 3)](https://padlet.com/julia_dumont/covid19-les-graphiques-nifnaf-puilc4nj7dwj/slideshow/wish/1367079090)

```python
pf.setvis('bokeh')
pf.setwhom('spf',reload=True)
pf.plot(which='cur_rea',where="Ile-de-France",option='sumall',when='22/03/2020:28/03/2021')
pf.setvis('matplotlib')
d3=pf.get(which='cur_rea',where="Ile-de-France",option='sumall',when='22/03/2020:28/03/2021')
b3=pf.plot(which='cur_rea',where="Ile-de-France",option='sumall',when='22/03/2020:28/03/2021')
```

```python
import numpy as np

t=["2020-04-09","2020-11-13","2021-03-28"]
td=[np.datetime64(x) for x in t]
y=d3[d3.date.isin(td)].cur_rea
ylabel=[str(int(b))+" ("+a+")" for a,b in zip(t,y)]

def annotate(ax, label, x, y, xytext):
    ax.annotate(label, xy=(x,y), 
                xytext=xytext, textcoords='offset points', 
                fontsize=15, 
                arrowprops={'arrowstyle': '-|>', 'color': 'black'})

# conditionally position labels
for label, x0, y0 in zip(ylabel, td, y):
    annotate(b3, label, x0, y0, (-45, -80))

b3.grid()
b3.figure.savefig("padlet_p3bis")
```
