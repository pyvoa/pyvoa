## France 2 - Journal de 20h du 15/04/2021 (padlet page 2)

```python
pf.setvis('matplotlib')
pf.setwhom('spf',reload=False)
pf.map(which='tot_dchosp',when='15/04/2021',mapoption='dense')
```

## Evolution du nombre de patients en réanimation en Ile de France depuis le 22/03/2020 (première vague) (padlet page 3)

```python
pf.setvis('bokeh')
pf.setwhom('spf',reload=False)
pf.plot(which='cur_rea',option='sumall',when='22/03/2020:28/03/2021')
```

```python
pf.setwhom('owid',reload=False)
pf.plot(which='cur_icu_patients',where='fra',when='22/03/2020:28/03/2021')
```

## Réception de doses vaccinales début 2021 (padlet page 5)
Note : pas d'info détaillée par type de vaccin dans les données SPF

```python
pf.setwhom('spf',reload=False)
pf.plot(which='tot_vacc1',option='sumall',when='01/02/2021:08/03/2021')
```

```python
pf.setwhom('spf',reload=False)
pf.plot(which='tot_vacc1',option=['normalize:pop100k','sumall'],when='01/02/2021:08/03/2021')
```

```python
pf.setwhom('owid',reload=False)
pf.plot(which='total_vaccinations',where='FRA',option='normalize:pop100k',when='01/02/2021:08/03/2021')
```
