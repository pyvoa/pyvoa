```python
from IPython.display import Markdown
import random
import pyvoa.tools as pt
pt.set_verbose_mode(1)
```

```python
import pyvoa.front as pf
print(pf.listvis())
```

```python
# Test rapide des bases et visu
for v in pf.listvis():
    if v=='folium':
        continue
    pf.setvis(v)
    display(Markdown('# Visu = '+v))
    for db in pf.listwhom():
        #if db in ['imed']:
        #    continue
        display(Markdown('## '+db))
        if db != pf.getwhom(None,False,False):
            pf.setwhom(db,reload=False)
        if type(pf.listwhere())==list:
            w=random.choices(pf.listwhere(),k=3)
            print(w)
            pf.plot(where=w)
        else:
            pf.plot()
```

```python
pf.setwhom('mpoxgh',reload=False) # not too many which to test
for v in pf.listvis():
    if v in ['folium','seaborn']:
        continue
    pf.setvis(v)
    display(Markdown('# Visu = '+v))
    for whi in pf.listwhich():
        for wha in pf.listwhat():
            lo=pf.listoption()
            lo.append('')
            for opt in lo:
                if opt=='nonneg' and whi[0:3]=='cur':
                    continue
                for plo in pf.listplot():
                    if plo in ['spiral','menulocation'] and vis=='matplotlib':
                        continue
                    if plo in ['versus']:
                        continue
                    if plo == 'yearly' and opt!='sumall':
                        continue
                    print(whi,wha,opt,plo,"plot")
                    pf.plot(which=whi,what=wha,option=opt,typeofplot=plo)
                for his in pf.listhist():
                    print(whi,wha,opt,his,"hist")
                    pf.hist(which=whi,what=wha,option=opt,typeofhist=his)
                if v=='seaborn':
                    continue # pas de map
                for mo in pf.listmapoption():
                    print(whi,wha,opt,'map',mo)
                    #for lt in pf.listtiles():
                    #    print(lt) â†’ no test of tiles currently
                    pf.map(which=whi,what=wha,option=opt,mapoption=mo) #,tile=lt)
```
